Here, I get 3 seperate dataframes, one for each participant, using their trial keys. I also filter the dataframes to include only the trial phase in which the participant is in control of the chair. 

```python
good_parts = ["1_rf_P8/", "1_fp_P10", "1_bd_P16", "1_nr_P25",
             "2_rf_P8/", "2_fp_P10", "2_bd_P16", "2_nr_P25"]
df = df[df['peopleName'].isin(good_parts)]
df
```

```python
df = df[df['trialPhase'] == 3] #filter out reset and crash phases
df
```

Filtering out unused columns

The \*_simple_df is used to create the input arrays for my models as well as for plotting the original values from the dataset (after I convert to numpy array cause I know how to handle those)

```python
df = df.drop(columns=['peopleTrialKey', 'calculated_vel', 'peopleName'])
df
```

Adding a columns of (x,y) coordinates using the position angle and converting my angular velocity from &deg;/sec to rad/sec because that is the observation state the environments accept for input.

```python
#https://sparkbyexamples.com/pandas/pandas-add-column-based-on-another-column/

df = df.assign(x_value=lambda x: np.cos(np.radians(x.currentPosRoll)))
df = df.assign(y_value=lambda y: np.sin(np.radians(y.currentPosRoll)))
df = df.assign(velRads=lambda rad: np.radians(rad.currentVelRoll))
df.iloc[1500:1510, :]
```

```python
print(np.cos(3.155823))
print(np.cos(np.radians(3.155823)))
```

```python
df['joystickX'] = df['joystickX'].apply(lambda x: x*-1) #multiply joystick by -1, per dataset instructions
df.iloc[1500:1510, :]
```

Converting to numpy arrays because I'm more familar with how to use these than I am with dataframes.

```python
obs_df = df.drop(columns=['seconds', 'trialPhase', 'currentPosRoll', 'joystickX', 'currentVelRoll'])
obs_df
```

```python
# act_df = df['joystickX']
# act_df
act_df = pd.DataFrame().assign(joystick=df['joystickX'])
act_df
```

```python
obs_arr = obs_df.to_numpy()
act_arr = act_df.to_numpy()
print(obs_arr.shape)
print(act_arr.shape)
```

```python
print(obs_arr[0])
print(act_arr[0])
```

```python
np.savez_compressed(
    "MARS_BC_expert_transitions",
    expert_actions=act_arr,
    expert_observations=obs_arr,
)
```
